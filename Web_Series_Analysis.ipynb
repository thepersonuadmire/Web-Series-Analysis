```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (adjust file name if needed)
df = pd.read_csv("/content/Web_Series_Dataset - Sheet1.csv")

# Preview the data
df.head()
```

```python
# Task 1: Trend of Web Series Released Per Year
yearly_counts = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker='o', linewidth=2)
plt.title('Trend of Web Series Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Web Series')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Task 2: Distribution of IMDB Ratings Across Streaming Platforms (Box Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Platform', y='IMDB Rating', data=df)
plt.title('Distribution of IMDB Ratings Across Platforms')
plt.xlabel('Platform')
plt.ylabel('IMDB Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Task 3: Top 5 Highest Rated Web Series (Horizontal Bar Chart)
top5 = df.sort_values(by='IMDB Rating', ascending=False).head(5)
plt.figure(figsize=(10, 6))
sns.barplot(x='IMDB Rating', y='Title', data=top5, palette='viridis')
plt.title('Top 5 Highest Rated Web Series')
plt.xlabel('IMDB Rating')
plt.ylabel('Web Series Title')
plt.tight_layout()
plt.show()

```

```python
# Task 4: Genre Distribution of Web Series (Bar Chart)
plt.figure(figsize=(10, 6))
genre_counts = df['Genre'].value_counts()
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='pastel')
plt.title('Genre Distribution of Web Series')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Task 5: Relationship Between Number of Episodes and IMDB Rating (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Episodes', y='IMDB Rating', data=df, hue='Platform', palette='tab10')
plt.title('Episodes vs IMDB Rating')
plt.xlabel('Number of Episodes')
plt.ylabel('IMDB Rating')
plt.legend(title='Platform', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Task 6: Platform-Wise Average IMDB Rating by Genre (Heatmap)

# Create pivot table
pivot = df.pivot_table(values='IMDB Rating', index='Genre', columns='Platform', aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(12, 7))
sns.heatmap(pivot, annot=True, cmap='coolwarm', fmt=".1f", linewidths=0.5)
plt.title('Average IMDB Rating by Genre and Platform')
plt.xlabel('Platform')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
# Task 7: Episodes Per Season Analysis

# Add new column: Episodes per Season
df['Episodes per Season'] = df['Episodes'] / df['Seasons']

# Plot bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Title', y='Episodes per Season', data=df, palette='mako')
plt.title('Average Episodes per Season for Each Web Series')
plt.xlabel('Web Series')
plt.ylabel('Episodes per Season')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Task 8: IMDB Rating vs Number of Seasons (Grouped by Platform)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Seasons', y='IMDB Rating', hue='Platform', palette='Set2')
plt.title('IMDB Rating vs Number of Seasons by Platform')
plt.xlabel('Number of Seasons')
plt.ylabel('IMDB Rating')
plt.legend(title='Platform', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Task 9: Most Consistent Genre (Lowest Rating Variance)

# Group by genre and calculate rating variance
variance_df = df.groupby('Genre')['IMDB Rating'].var().reset_index().sort_values(by='IMDB Rating')

# Plot bar chart of variance
plt.figure(figsize=(10, 6))
sns.barplot(x='Genre', y='IMDB Rating', data=variance_df, palette='flare')
plt.title('Rating Variance per Genre (Lower = More Consistent)')
plt.xlabel('Genre')
plt.ylabel('IMDB Rating Variance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Task 10: Above-Average Rated Series in Each Genre

# Calculate average rating per genre
genre_avg = df.groupby('Genre')['IMDB Rating'].mean().reset_index()
genre_avg.columns = ['Genre', 'Genre Avg Rating']

# Merge with original dataframe
merged_df = pd.merge(df, genre_avg, on='Genre')

# Filter series with rating above genre average
above_avg = merged_df[merged_df['IMDB Rating'] > merged_df['Genre Avg Rating']]

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Title', y='IMDB Rating', data=above_avg, palette='spring')
plt.title('Web Series with Above-Average Ratings in Their Genre')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
